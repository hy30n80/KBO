## 투수, 타자 공통으로 갖고 있는 10개의 Feature 만 추출해서 Preprocessed_Feature 추출

```python
pitcher = pd.read_csv("./Final Dataset/Pitcher Feature.csv")
batter = pd.read_csv("./Final Dataset/Batter Feature.csv")
edge = pd.read_csv("./Final Dataset/Edge Feature.csv")
```

```python
pitcher.columns
```

```python
batter
```

```python
p_pitcher = pitcher[['player_id', 'Name', 'play_type', 'team', 'TBF', 'AVG', 'OBP', 'SLG', 'K%', 'BB%']]
```

```python
p_pitcher
```

```python
batter.columns
```

```python
p_batter = batter[['player_id', 'Name', 'play_type', 'team', 'PA', 'AVG', 'OBP', 'SLG', 'K%', 'BB%']]
```

```python
p_batter
```

```python
# 새로운 컬럼 추가: 'bat_side'
def get_bat_side(play_type):
    if "우타" in play_type:
        return "Right"
    elif "좌타" in play_type:
        return "Left"
    elif "양타" in play_type:
        return "Both"
    else:
        return "Unknown"

def get_pit_side(play_type):
    if "우투" in play_type:
        return "Right"
    elif "우언" in play_type:
        return "Right"

    elif "좌투" in play_type:
        return "Left"
    elif "좌언" in play_type:
        return "Left"
    
    elif "양투" in play_type:
        return "Both"
        
    else:
        return "Unknown"

p_batter["Bat_side"] = p_batter["play_type"].apply(get_bat_side)
p_pitcher["Pit_side"] = p_pitcher["play_type"].apply(get_pit_side)
```

```python
p_batter = p_batter.rename(columns={
    "player_id": "ID",
    "team": "Team"
})

p_pitcher = p_pitcher.rename(columns={
    "player_id": "ID",
    "team": "Team"
})
```

```python
p_pitcher.isnull().sum()
```

```python
p_batter = p_batter.drop(columns=['play_type'])
p_pitcher = p_pitcher.drop(columns=['play_type'])
```

```python
p_pitcher.to_csv("./Final Dataset/Preprocessed_Pitcher_Feature.csv")
```

```python
p_batter.to_csv("./Final Dataset/Preprocessed_Batter_Feature.csv")
```
